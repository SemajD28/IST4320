```python
# question 2 : Here is a list of groceries. The user needs help extracting the text "apples" and placing the text "apples" into the sentence below.
#  Please use regular expressions. ( regex ) 
import re

random_text = "Milk, eggs, bread, butter, cheese, fresh fruits (e.g., apples, bananas), fresh vegetables (e.g., lettuce, tomatoes), rice, pasta, canned tomatoes, chicken breasts or thighs, ground beef or turkey, cooking oil (e.g., olive oil), spices (e.g., salt, pepper), snacks (e.g., chips, nuts)"

# Use regular expression to find "apples"
match = re.search(r'\bapples\b', random_text)

if match:
    # Extract the matched text
    apples_text = match.group(0)
    
    # Sentence to insert "apples"
    new_sentence = "I need to buy {} today.".format(apples_text)

    print(new_sentence)
else:
    print("The text 'apples' was not found in the given sentence.")

#What is the starting and ending position of the text "apples" in random_text? Paste below üëá 
import re

random_text = "Milk, eggs, bread, butter, cheese, fresh fruits (e.g., apples, bananas), fresh vegetables (e.g., lettuce, tomatoes), rice, pasta, canned tomatoes, chicken breasts or thighs, ground beef or turkey, cooking oil (e.g., olive oil), spices (e.g., salt, pepper), snacks (e.g., chips, nuts)"

# Use regular expression to find "apples"
match = re.search(r'\bapples\b', random_text)

if match:
    # Get starting and ending positions of the match
    start_position = match.start()
    end_position = match.end()

    print("Starting position:", start_position)
    print("Ending position:", end_position)
else:
    print("The text 'apples' was not found in the given sentence.")     
      
```

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

# URL of the website
site = "https://littlespoonfarm.com/apple-pie-recipe/"

# Fetch the HTML content of the website
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site, headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page, 'html.parser')

# Find all 'ul' and 'ol' elements containing the recipe ingredients
ingredient_lists = soup.find_all(['ul', 'ol'])

# Extract the text content of each ingredient list
recipe_ingredients = []
for ingredient_list in ingredient_lists:
    ingredients = ingredient_list.find_all('li')
    for ingredient in ingredients:
        recipe_ingredients.append(ingredient.text.strip())

# Print the recipe ingredients
print("Recipe Ingredients:")
for ingredient in recipe_ingredients:
    print(ingredient)



```
